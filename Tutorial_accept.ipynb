## Normalized maginal densities

```python
n1 = 5
n2 = 10
n3 = 30
ret = compute_pdf_scoring(data, h, flight_numbers,
                          n_low_res_steps=n1,
                          n_high_res_steps=n2,
                          n_low_res_steps2=n3,
                          test_flights = test_flights_names)
```

The previous function returns, in this order,
* the arrays of scaled marginal likelihoods at each point of the test flights,
* the MML scores of the test flights for each variable,
* the mean MML for each test flight,
* the trained fastkde estimators at each bin, and so on...

```python
pdf_per_test_flight, scores_per_dim_list, mean_scores, pdfobj_per_h, h_train, train, h_test_list, test_list, h_windows, var_bounds, total_training_time, n_training_points = ret
```

```python
plot_heatmap('test_heatmap_pdf.png', pdfobj_per_h, h, data, h_windows)
```

You can find the MML scores in the scores_per_dim_list returned:

```python
print("\nSCORES:\n")
for scores_per_dim_k, mean_score in zip(scores_per_dim_list, mean_scores):
    for i, score in enumerate(scores_per_dim_k):
        print("Dimension %i = %.2f" % (i, score))
    print("\nMean score = %.2f \n" % mean_score)
```
